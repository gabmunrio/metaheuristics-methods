```python
#!/usr/bin/env python

'''
pso.py

A simple implementation of the Particle Swarm Optimisation Algorithm.
Uses Numpy for matrix operations.
Pradeep Gowda 2009-03-16
'''

import math
import numpy as np
import matplotlib.pyplot as plt
import random

from numpy import array
from math import sin, sqrt
from matplotlib import animation


class Particle:
    pass


waldo_positions = [[0.625,7.708333333],[4.944444444,6.569444444], [5.430555556,6.402777778], [5.902777778,6.083333333],
                   [5.430555556,5.444444444], [4.791666667,5.444444444], [5.305555556,5.125], [3.666666667,5.291666667],
                   [6.069444444,4.638888889], [3.513888889,4.319444444], [2.388888889,4.486111111], [1.763888889,4.333333333],
                   [2.083333333,3.847222222], [1.125,2.569444444], [0.805555556,2.25], [1.444444444,2.097222222],
                   [0.638888889,1.777777778], [0.958333333,1.277777778], [0.958333333,0.972222222], [2.083333333,0.805555556],
                   [2.388888889,2.729166667], [2.166666667,2.25], [2.486111111,2.25], [3.347222222,2.569444444],
                   [3.513888889,2.083333333], [3.041666667,1.916666667], [4.305555556,2.888888889], [5.263888889,2.729166667],
                   [5.75,2.25], [5.597222222,1.930555556], [4.625,1.277777778], [5.263888889,0.819444444],
                   [5.416666667,0.333333333], [7.666666667,7.361111111], [6.861111111,6.736111111], [8.458333333,6.722222222],
                   [10.375,7.194444444], [12.44444444,7.694444444], [10.375,6.430555556], [11.16666667,6.402777778],
                   [11.33333333,6.083333333], [9.888888889,5.777777778], [9.25,5.291666667], [9.958333333,4.791666667],
                   [9.097222222,4.652777778], [7.944444444,5.305555556], [7.763888889,5], [8,4.791666667],
                   [7.333333333,5.138888889], [6.694444444,4.972222222], [7.166666667,4.5], [7.486111111,4.006944444],
                   [11.01388889,4.819444444], [11.80555556,4.652777778], [12.27777778,4.805555556], [9.416666667,3.694444444],
                   [9.25,3.222222222], [11.48611111,3.541666667], [11.95833333,3.541666667], [7.986111111,2.430555556],
                   [6.694444444,1.625], [12.125,2.909722222], [11.51388889,2.756944444], [11.30555556,2.555555556],
                   [10.72222222,1.930555556], [11.16666667,1.791666667], [12.44444444,1.625], [12.125,1.291666667]]

# Function which calculates the euclidean distance between two points
def euclideanDistance(index1, index2):
    sum = 0.0

    v1 = waldo_positions[index1]
    v2 = waldo_positions[index2]

    for coord1, coord2 in zip(v1, v2):
        sum += pow((coord1 - coord2), 2)

    return math.sqrt(sum)


# Function that evaluates the total length of a path
def calculate_fitness(path):
    # Here tour cost refers to the sum of the euclidean distance between
    # consecutive points starting from first element
    totalDistance = 0.0
    size = len(path)
    index = 0
    while index < size:
        # select the consecutive point for calculating the segment length
        if index == size - 1:
            # This is because in order to complete the 'tour' we need to reach
            # the starting point
            point2 = path[0]
        else:  # select the next point
            point2 = path[index + 1]

        totalDistance += euclideanDistance(path[index], point2)
        index = index + 1

    return totalDistance


# Function that defines the velocity
def define_velocity(pr1, pr2, pr3, p, gbest):
    if pr1 > pr2:
        if pr1 > pr3:
            return p.params
        else:
            return gbest.params
    elif pr2 > pr3:
        return p.best
    else:
        return gbest.params

'''
# Function that updates the velocity:
def update_velocity(p, velocity):
    v_fitness = calculate_fitness(velocity)
    p_fitness = calculate_fitness(p)
    
    v_result = p
    if v_fitness < p_fitness:
        v_result = velocity
    return v_result

'''

# Function that updates the velocity
# We apply a change to move from orig to dest
def update_velocity(orig, dest):
    # We obtain two randoms number to get a subset
    tour_size = len(dest)
    start = random.randint(0, tour_size - 1)
    end = random.randint(0, tour_size - 1)
    if end == tour_size:
        end = tour_size - 1
    
    # We copy the subset to the childs
    child1 = list(dest[start:end])

    # We iterate over the cities to add it or not to the childs
    for city in orig:
        if city not in child1:
            child1.append(city)

    return child1

'''
def update_velocity(orig, dest):
    # We obtain two randoms number to get a subset
    tour_size = len(dest)
    start = random.randint(0, tour_size - 1)
    end = start + 2
    if end == tour_size:
        end = 0

    # We copy the subset to the childs
    child1 = set()
    child1 |= set(dest[start:end])
    
    # We iterate over the cities to add it or not to the childs
    for city in orig:
        if city not in child1:
            child1.add(city)

    child1 = list(child1)

    return child1
'''

# Initial parameters

iter_max = 10000 # Stopping criteria
pop_size = 20  # Size of the population
i = 0

# initialize the particles
particles = []
for i in range(pop_size):
    p = Particle()
    len_waldo_positions = len(waldo_positions)
    p.params = np.random.permutation(len_waldo_positions)
    p.params= list(p.params)
    p.best = p.params
    p.fitness = 999999999
    p.v = 0.0
    particles.append(p)

# Initialize the probabilities to follow it's own way, pbest or gbest    
# The three probabilites sum equals 1
pr1 = 0.9  # Own way
pr2 = 0.05  # pBest
pr3 = 0.05  # gBest

# let the first particle be the global best
gbest = particles[0]
err = 999999999

while i < iter_max:
    for p in particles:
        fitness = calculate_fitness(p.params)

        if fitness < p.fitness:
            p.fitness = fitness
            p.best = p.params
            
        if fitness < gbest.fitness:
            gbest = p
            # Let's draw the solution
            plt.clf()
            result_x = []
            result_y = []
            for index in gbest.params:
                coordinates = waldo_positions[index]
                result_x.append(coordinates[0])
                result_y.append(coordinates[1])
            plt.plot(result_x, result_y, 'ro')

            # We draw the arrows
            index = 1
            total_length = len(result_x)

            while index < total_length:
                origin_x = result_x[index-1]
                origin_y = result_y[index-1]
                dest_x = (result_x[index] - result_x[index-1])
                dest_y = (result_y[index] - result_y[index-1])

                ax = plt.axes()
                ax.arrow(origin_x, origin_y, dest_x, dest_y)
                index = index + 1 

            # plt.show()
            plt.draw()
        
        # print the temporary values
        # print "{0} | {1}".format(fitness, gbest.fitness)
        
        # We need to define the function to define the velocity
        p_velocity = define_velocity(pr1, pr2, pr3, p, gbest)
        
        actual_path = list(p.params)
        guiding_path = list(p_velocity)
        
        # Update the velocity
        p.params = update_velocity(actual_path, guiding_path)
    
    
    # print the temporary values
    print "\n{0}".format(gbest.fitness)
    
    # Update iterator index
    i += 1
    
    # Update probabilites
    pr1 = pr1 * 0.90
    pr2 = pr2 * 1.01
    pr3 = 1 - (pr1 + pr2)

    
# Uncomment to print particles
# for p in particles:
#    print 'params: %s, fitness: %s, best: %s' % (p.params, p.fitness, p.best)
```
