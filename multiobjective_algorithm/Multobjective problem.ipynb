```python
import math
import random
import matplotlib.pyplot as plt

# Our problem is to optimize the parameters to build a cone with some constraints

# Right circular cone:

# r = base radius  - Input variable number 1, value between 0,10
# h = height  - Input variable number 2, value between 0,20

# s = slant height
# V = volume  - One constraint, V > 200
# B = base area
# S = lateral surface area  - One objective, minimize S
# T = total area  - One objective, minimize T

# Formulas:
# s = sqrt(r^2 + h^2)
# V = pi/3 * r^r * h
# B = pi * r^2
# S = pi * r * s
# T = B + S = pi * r * (r + s)

# Some constant
PI = math.pi

# Function to calculate the pareto frontier of the results
def pareto_frontier(Xs, Ys, maxX = True, maxY = True):
# Sort the list in either ascending or descending order of X
    myList = sorted([[Xs[i], Ys[i]] for i in range(len(Xs))], reverse=maxX)
# Start the Pareto frontier with the first value in the sorted list
    p_front = [myList[0]]    
# Loop through the sorted list
    for pair in myList[1:]:
        if maxY: 
            if pair[1] >= p_front[-1][1]: # Look for higher values of Y…
                p_front.append(pair) # … and add them to the Pareto frontier
        else:
            if pair[1] <= p_front[-1][1]: # Look for lower values of Y…
                p_front.append(pair) # … and add them to the Pareto frontier
# Turn resulting pairs back into a list of Xs and Ys
    p_frontX = [pair[0] for pair in p_front]
    p_frontY = [pair[1] for pair in p_front]
    return p_frontX, p_frontY

# Class to store the objectives to apply to our population
class Objectives(object):
    
    # To calculate the two objectives functions, we need auxiliar functions
    # Function to calculate the slant height: s = sqrt(r^2 + h^2)
    def calculate_s(self, r, h):
        first_comp = r**2
        second_comp = h**2
        
        return math.sqrt(first_comp + second_comp)
    
    # Function to calculate the base area: B = pi * r^2
    def calculate_B(self, r):
        second_comp = r**2
        
        return PI * second_comp
    
    # Function to calculate the volume: V = pi/3 * r^r * h. Is our constraint.
    def calculate_V(self, r, h):
        first_comp = PI/3
        second_comp = r**r
        third_comp = h
        
        return first_comp * second_comp * third_comp
    
    # One objective is to minimize the lateral surface area: S = pi * r * s
    def calculate_S(self, element):
        r = element[0]
        h = element[1]
        
        # We need to check our constraint
        volume = self.calculate_V(r, h)
        if volume <= 200:
            # We return a big number, to penalize
            return 999999
        
        s = self.calculate_s(r, h)
        S = PI * r * s
        
        return S
    
    # One objective is to minimize the total area: T = B + S = pi * r * (r + s)
    def calculate_T(self, element):
        r = element[0]
        h = element[1]
        
        # We need to check our constraint
        volume = self.calculate_V(r, h)
        if volume <= 200:
            # We return a big number, to penalize
            return 999999
        
        s = self.calculate_s(r, h)
        
        first_times = PI * r
        second_add = r + s
        
        return first_times * second_add

    
# Pareto domination
def pareto_domination(a, b, objectives):
    # We get an instance of the class with the functions defined
    objective_instance = Objectives()
    result = False
    for objective in objectives:
        objective_func = getattr(Objectives, "{0}".format(objective))
        results_a = objective_func(objective_instance, a)
        results_b = objective_func(objective_instance, b)
        
        # We store the values for the objective functions
        if objective is 'calculate_S':
            a[2] = results_a
            b[2] = results_b
        
        if objective is 'calculate_T':
            a[3] = results_a
            b[3] = results_b
        
        # We need to minimize, that's because we use '<'
        if results_a < results_b:
            result = True
        else:
            return False
    return result


# Pareto domination Binary tournament selection
# Our population will be 10 individuals
population = []
population_size = 1000000
for i in range(population_size):
    element_r = random.uniform(0, 10)
    element_h = random.uniform(0, 20)
    # We want to store the two values for the objectives
    element_S = 0
    element_T = 0
    element = [element_r, element_h, element_S, element_T]
    population.append(element)

# Our list of objectives
objectives = ['calculate_S', 'calculate_T']

# We will save the solutions, to compare which one minimize each objective function per separate
solutions = []

# We assign random parameter to the best individual too
best_r = random.uniform(0.0, 10.0)
best_h = random.uniform(0.0, 20.0)
best_S = 0
best_T = 0
best_individual = [best_r, best_h, best_S, best_T]

for element in population:
    pareto_domination_result = pareto_domination(best_individual, element, objectives)
    # print "best individual - {0} | best possible - {1}\n".format(best_individual, element)
    # If false, We need to update the best individual
    if not pareto_domination_result:
        best_individual = element
        solutions.append(element)

# print "Final results: best individual - {0}\n".format(best_individual)
# print "Solutions: {0}".format(solutions)

# Let's draw the population to compare with the solution space
population_x = []
population_y = []
for individual in population:
    population_x.append(individual[0])
    population_y.append(individual[1])
plt.plot(population_x, population_y, 'ro')
# plt.scatter(result_x, result_y)
# Let's draw the solution space
sol_x = []
sol_y = []
for solution in solutions:
    sol_x.append(solution[0])
    sol_y.append(solution[1])
plt.plot(sol_x, sol_y, 'bo')
# plt.scatter(result_x, result_y)
p_front = pareto_frontier(sol_x, sol_y) 
plt.plot(p_front[0], p_front[1])
plt.show()
```
